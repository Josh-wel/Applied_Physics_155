```python
import numpy as np
import matplotlib as plt
```

```python
#Constants value
a_1 = 15.8
a_2 = 18.3
a_3 = 0.714
a_4 = 23.2

#Binding Energy values
A = 58
Z = 28

def semi_empirical_mass(A, Z):
    if A % 2 == 1: 
        a_5 = 0
    elif Z % 2 == 0:
        a_5 = 12
    else:
        a_5 = -12
    return (a_1 * A) - (a_2 * A ** (2/3)) - (a_3 * ((Z ** 2) / (A ** (1/3)))) - (a_4 *((A - 2 * Z)**2) / A) + (a_5 / (A ** (1/2)))

print('semi empirical mass =' , semi_empirical_mass(A, Z))
```

```python
def binding_energy_per_nucleon(A,Z):
    return (semi_empirical_mass(A,Z) / A)

print('binding energy per nucleon = ', binding_energy_per_nucleon(A, Z))
```

```python
def largest_binding_energy_per_nucleon(Z):
    A = Z
    A_most_stable = 0
    max_binding_energy_per_nucleon = 0
    for A in range (Z, 3 * Z + 1):
        T = binding_energy_per_nucleon(A, Z)
        if T >= max_binding_energy_per_nucleon:
            max_binding_energy_per_nucleon = T
            A_most_stable = A
    return A_most_stable, max_binding_energy_per_nucleon    

print(largest_binding_energy_per_nucleon(Z))
```

```python
print("Z ", "A", " largest binding energy per nucleon")
max_binding_energy_per_nucleon = 0
A_most_stable = 0
Z_max = 0
for Z in range(1, 101):
    A, Max = largest_binding_energy_per_nucleon(Z)
    if Max >= max_binding_energy_per_nucleon:
        max_binding_energy_per_nucleon = Max
        A_most_stable = A
        Z_max = Z
    print(Z , largest_binding_energy_per_nucleon(Z))
    
print("Z=", Z_max, "A_most_stable=", A_most_stable,"maximum binding energy per nucleon=" , max_binding_energy_per_nucleon)
```
